```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
```

```python
runge = lambda x : 1 / (1 + 25 * x ** 2) # Runge function : 1 / (1 + 25x^2)
```

```python
x_new = np.linspace(-1, 1, 1000)
```

```python
fig = plt.figure(figsize=(10, 8))
plt.plot(x_new, runge(x_new), 'b')
L = ['Runge function']
for n in np.arange(5, 25, 5):
  x = np.linspace(-1, 1, n)
  f = runge(x)
  y = lagrange(x, f)
  plt.plot(x_new, y(x_new))
  L.append('Interpolating Polynomial of degree {}'.format(n))
plt.legend(L)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
