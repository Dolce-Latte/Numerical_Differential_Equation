```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2, 3, 4]
f = [1, 3, 2, 5, 0]
# x = [0, 1, 2, 3]
# f = [1, 3, 2, 5]
```

```python
y = CubicSpline(x, f, bc_type='natural')
```

```python
x_new = np.linspace(0, 4, 1000)
y_new = y(x_new)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, f, 'ro')
plt.title('Cubic Spline Interpolation(natural)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
