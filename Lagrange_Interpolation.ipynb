```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2] #nodes points
f = [1, 3, 2] #values
```

```python
#get the Lagrange polynomial functions
L0_coeff = [1, -1.5, 0.5]
L1_coeff = [0, 2, -1]
L2_coeff = [0, -0.5, 0.5]
L0 = poly.Polynomial(L0_coeff)
L1 = poly.Polynomial(L1_coeff)
L2 = poly.Polynomial(L2_coeff)
```

```python
#visualize the Lagrange polynomials
x_new = np.arange(-1.1, 3.1, 0.1)
fig = plt.figure(figsize=(10, 8))
plt.plot(x_new, L0(x_new), 'b', label='L0')
plt.plot(x_new, L1(x_new), 'r', label='L1')
plt.plot(x_new, L2(x_new), 'g', label='L2')
plt.plot(x, np.ones(len(x)), 'ko', x, np.zeros(len(x)), 'ko')
plt.title('Lagrange Basis Polynomials')
plt.grid()
plt.legend()
plt.show()
```

```python
# interpolation polynomial
P = f[0] * L0 + f[1] * L1 + f[2] * L2
```

```python
fig = plt.figure(figsize=(10, 8))
plt.plot(x_new, P(x_new), 'b', x, f, 'ro')
plt.title('Lagrange Interpolating Polynomial')
plt.grid()
plt.show()
```

```python
from scipy.interpolate import lagrange
```

```python
p = lagrange(x, f)
```

```python
fig = plt.figure(figsize=(10, 8))
plt.plot(x_new, p(x_new), 'b', x, f, 'ro')
plt.title('Lagrange Interpolating Polynomial(Scipy)')
plt.grid()
plt.show()
```
